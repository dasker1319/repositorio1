```python
# Inventario
inventario = [
    {"nombre": "manzana", "precio": 2000, "stock": 50},
    {"nombre": "plátano", "precio": 1000, "stock": 100},
    {"nombre": "leche", "precio": 4500, "stock": 30}
]
# Entrada del usuario
nombre = input("Ingresa el nombre del producto: ")
# Buscar y mostrar el producto
for producto in inventario:
    if producto["nombre"] == nombre:
        print(producto)
        break
else:
    print("Producto no encontrado.")
```

```python
ventas = [# Lista de los productos
    ("Laptop", "Electrónica", 1500),
    ("Camisa", "Ropa", 80),
    ("Mouse", "Electrónica", 50),
    ("Pantalón", "Ropa", 120),
    ("Teclado", "Electrónica", 70)
]
# Diccionario para almacenar
total_ventas = {}
# Recorremos las ventas y sumamos el monto por categoría
for producto, categoria, monto in ventas:
    if categoria in total_ventas:
        total_ventas[categoria] += monto
    else:
        total_ventas[categoria] = monto
print(total_ventas)# Mostrar el total de ventas por categoría
```

```python
# Diccionario con productos
productos = {
    101: "Leche",
    102: "Pan",
    103: "Huevos"
}
# Diccionario con precios
precios = {
    101: 4500,
    102: 2500,
    103: 8000
}
# Lista de ventas del día
ventas_del_dia = [
    (101, 5),  # Se vendieron 5 leches
    (102, 10), # Se vendieron 10 panes
    (101, 3),  # Se vendieron 3 leches mas
    (103, 2)   # Se vendieron 2 huevos
]
# Calcular el ingreso total del día
ingreso_total = 0
for producto_id, cantidad in ventas_del_dia:
    precio = precios[producto_id]  # Obtener el precio del producto
    ingreso_total += precio * cantidad  # Sumar el ingreso de esa venta
# Mostrar el ingreso total
print("Ingreso total del día:", ingreso_total)
```
