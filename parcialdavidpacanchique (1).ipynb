```python
#2nodo para lista enlazada circular
class Nodo:
    def __init__(self, nombre, gesto):
        self.nombre = nombre
        self.gesto = gesto
        self.sig = None

#datos
nombres = [
    "Ana","Luis","Carlos","María","Pedro",
    "Lucía","Andrés","Camila","Sofía","Juan",
    "Valentina","David","Paola","Mateo","Sara",
    "Felipe","Laura","Esteban","Daniela","Tú"
]

gestos = [
    "mirando el reloj",
    "revisando el celular nervioso",
    "dando golpecitos con el pie",
    "acomodando la mochila rápido",
    "suspiro constante",
    "cambiando de peso de un pie a otro",
    "consultando la hora en el reloj",
    "mirando hacia la puerta varias veces",
    "apretando los labios",
    "hojeando papeles con prisa",
    "mirando la bandeja del barista",
    "golpeando suavemente la mesa",
    "moviendo la cabeza con impaciencia",
    "ordenando monedas en la mano",
    "mordiéndose el labio",
    "observando repetidamente la cafetera",
    "mirando la fila con impaciencia",
    "respirando profundo varias veces",
    "manteniendo la mirada fija en el barista",
    "alzando la voz y empujando (grosero)"
]

#crear nodos y enlazar circularmente
nodos = [Nodo(nombres[i], gestos[i]) for i in range(len(nombres))]
for i in range(len(nodos)):
    nodos[i].sig = nodos[(i+1) % len(nodos)]

#clasificar: normales y groseros
lista = []
groseros = []
act = nodos[0]
for _ in range(len(nodos)):
    if "grosero" in act.gesto:
        groseros.append(act)
    else:
        lista.append(act)
    act = act.sig

#orden final
final = lista + groseros

#imprimir
for i, persona in enumerate(final, 1):
    print(f"{i}. {persona.nombre} - {persona.gesto}")
```

```python
#7.organigrama de una empresa
#cada empleado se guarda como un diccionario con su nombre, cargo y lista de hijos
def crear_empleado(nombre, cargo):
    return {"nombre": nombre, "cargo": cargo, "hijos": []}

#busca al supervisor y le agrega un nuevo empleado
def agregar(arbol, supervisor, nombre, cargo):
    if arbol["nombre"] == supervisor:
        arbol["hijos"].append(crear_empleado(nombre, cargo))
        return True
    #si no es el supervisor, busca entre los hijos
    for hijo in arbol["hijos"]:
        if agregar(hijo, supervisor, nombre, cargo):
            return True
    return False

#muestra la jerarquía de la empresa
def mostrar(arbol, nivel=0):
    print("   " * nivel + "- " + arbol["nombre"] + " (" + arbol["cargo"] + ")")
    for hijo in arbol["hijos"]:
        mostrar(hijo, nivel + 1)

#crear al jefe principal (ceo)
empresa = crear_empleado("Laura", "CEO")

#agregar empleados bajo sus supervisores
agregar(empresa, "Laura", "Carlos", "Gerente de Ventas")
agregar(empresa, "Laura", "Ana", "Gerente de Finanzas")
agregar(empresa, "Carlos", "Luis", "Ejecutivo de Ventas")
agregar(empresa, "Ana", "Sofía", "Analista Financiera")

#mostrar toda la estructura
print("Organigrama de la Empresa:\n")
mostrar(empresa)
```
